**Ayuda Como generar los datos**


```python
import numpy as np
import matplotlib.pyplot as plt

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N ,[0.4, 0.5, 0.1])  # Por ejemplo, 40% masculino, 50% femenino, 10% no especificado)
    # Generar edades con distribución normal
    edad =np.random.normal(44, 6, N)
    edad=np.clip(np.around(edad),28,60)
    # Generar alturas, pesos y otras características
    altura = np.random.normal(180, 10, N)
    altura = np.clip(np.around(altura, decimals=1), 150.0, 210.0)

    peso = np.random.normal(80, 15, N)
    peso = np.clip(np.around(peso, decimals=1), 40.0, 130.0)

    horas_trabajo_estudio = np.random.normal(10, 2, N)
    horas_trabajo_estudio = np.clip(np.around(horas_trabajo_estudio, decimals=1), 4.0, 12.0)

    horas_ejercicio = np.random.normal(1, 0.5, N)
    horas_ejercicio = np.clip(np.around(horas_ejercicio, decimals=1), 0, 2)

    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    return datos



# Solicitar al usuario la cantidad N de datos a cargar
N = 10

# Generar datos aleatorios para N personas con distribución normal y límites específicos
datos = generar_datos(N)
#print (datos)
# extraer todas las columnas exepto la columna genero
columnas_sin_genero=datos[:,1:].astype(float)
# calcular la media a lo largo del eje 0(media de cada columna)
media_sin_genero=np.mean(columnas_sin_genero,axis=0)
print("medias de todas las columnas sin genero:")
print(media_sin_genero)
#graficar la muestra o ditribucion de los datos de peso, altura en graficas separadas . indicar el grado de variabilidad de sus datos
#obtener solo la columan de peso y atura
peso = datos[:,3].astype(float)
alturas = datos[:,2].astype(float)
edades = datos[:,1].astype(float)

#crear grafica de disribucion separados  para  peso y altura
plt.figure(figsize=(8,6))

#garficos de edades
plt.subplot(3,1,1)
plt.hist(edades, bins=20,edgecolor='black')
plt.xlabel('edades')
plt.ylabel('frecuencia')
plt.title('distribucion de edades')
std_edades= np.std(edades)
print("desviacion estandar de  edades:",std_edades)

#garficos de peso
plt.subplot(3,1,2)
plt.hist(peso, bins=20,edgecolor='black')
plt.xlabel('peso')
plt.ylabel('frecuencia')
plt.title('distribucion de peso')
std_peso= np.std(peso)
print("desviacion estandar de  peso:",std_peso)

#garficos de altura
plt.subplot(3,1,3)
plt.hist(alturas, bins=20,edgecolor='black')
plt.xlabel('alturas')
plt.ylabel('frecuencia')
plt.title('distribucion de alturas')
std_alturas= np.std(alturas)
print("desviacion estandar de  alturas:",std_alturas)
#ajustar espacio entre subplots
plt.tight_layout()
#mostrar los graficos
plt.show()


```

```python
#se define una función para generar y limpiar los outliers
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 0.5 * iqr
    upper_bound = q3 + 0.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Llamamos a la función para cada columna numérica donde desees detectar outliers
columns_to_check = ["Age"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df, col)
    outliers_list.append(outliers)

# Concatenamos todos los outliers encontrados en un solo DataFrame
outliers_df = pd.concat(outliers_list)

# Eliminamos los outliers del conjunto de datos original
df_cleaned1 = df.drop(outliers_df.index)

# Imprimimos la cantidad de outliers detectados
print("Número de outliers detectados:", len(outliers_df))
print(outliers)
```
